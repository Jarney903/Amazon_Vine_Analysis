```python
# Filter the data and create a new DataFrame or table 
# to retrieve all the rows where the total_votes count is equal to or greater than 20 
# to pick reviews that are more likely to be helpful and to avoid having division by zero errors later on.

df_filter1 = df[df["total_votes"] >= 20]
df_filter1.head()
```

```python
# Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table 
# to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.

df_filter1['percent_helpful']= df_filter1.apply(lambda row: row.helpful_votes / row.total_votes *100, axis=1)
df_filter1.head()
tot1 = len(df_filter1)
print(tot1)
```

```python
df_filter2 = df_filter1[df_filter1["percent_helpful"] >= 50]
df_filter2.head(50)
tot = len(df_filter2)
print(tot)
```

```python
# Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves 
# all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.

df_vine_yes = df_filter2[df_filter2["vine"] == 'Y']
df_vine_yes.head(20)
```

```python
# Repeat Step 3, but this time retrieve 
# all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.
df_vine_no = df_filter2[df_filter2["vine"] == 'N']
df_vine_no.head(20)
```

```python
## PAID -> vine = 'Y'
# Determine the total number of reviews, =7
y_reviews = len(df_vine_yes)
print(f"the number of reviews from vine subscribers is {y_reviews}")

# the number of 5-star reviews, = 0
five_reviews = len(df_vine_yes[df_vine_yes['star_rating'] == 5 ])
print(f"the number of 5-start reviews from vine subscribers is {five_reviews}")

# and the percentage of 5-star reviews
five_reviews_percent = five_reviews / y_reviews * 100
print(f"the number of percentage of 5-start reviews from vine subscribers is {five_reviews_percent} %")
```

```python
## UnPAID -> vine = 'N'
# Determine the total number of reviews, 
# the number of 5-star reviews,
# and the percentage of 5-star reviews
n_reviews = len(df_vine_no)
print(f"the number of reviews from non-vine subscribers is {n_reviews}")

# the number of 5-star reviews, = 0
five_reviews_n = len(df_vine_no[df_vine_no['star_rating'] == 5.0 ])
print(f"the number of 5-start reviews from non-vine subscribers is {five_reviews_n}")

# and the percentage of 5-star reviews
five_reviews_percent_n = five_reviews_n / n_reviews * 100
print(f"the number of percentage of 5-start reviews from non-vine subscribers is {five_reviews_percent_n} %")
```
